Funciones

```python
# se define la función 
def funcion (x):
  f=pow(x,2)+3*x-2
  return(f)
#se evalua la función
v=int(input("Escriba el valor de x...:"))
print("la función evaluada en el punto x=",v,"es igual a=",funcion (v))
```

Otra forma de definir una función 

```python
#calculo simbolico
f=input("ingrese la función en terminos de x...")
x=float(input("ingrese el valor de x"))
#eval evalua la función en el ultimo valor concido de x
y=eval(f)
print("la función evaluada en el punto x=",x,"es igual a=",format(y))
```

Gráficas

```python
#Librerias
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

vinf=input("Ingrese el valor inferior de x")
vsup=input("Ingrese el valor superior de x")
inc=float(input("ingrese el incremento"))

#se genera el vector
x=np.arange(vinf,vsup,inc)
y=np.sin(x)
z=np.cos(x)
y=pow(x,2)+1
plt.title("Hola Gráfica")
plt.xlabel("Eje x")
plt.ylabel("Eje Y")
plt.grid()
plt.axis([-100,100,-50,50])
plt.plot(y)
plt.plot(z)
```
