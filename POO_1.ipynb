Desarrollar un programa orientado a objetos que permita cálcular para un rectángulo, su área, su perimetro y el valor de su diagonal 

```python
class Rectangulo:
    def _init_(self,ancho,alto):
        self.ancho=ancho
        self.alto=alto
    def area(self):
        area=self.alto*self.ancho
        return area
   def perimetro(self):
       perimetro=(self.alto*2)+(self.ancho*2)
       return perimetro
r1=Rectangulo(4,2)
area=r1.area()
perimetro=r1.perimetro()
print("El area es:", area)
print("El perimetro es ", perimetro)
```
