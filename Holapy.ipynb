otra forma

```python

import numpy as np
a=float(input("ingrese el valor de a...:"))
b=float(input("ingrese el valor de b...:"))
c=a+b
d=a-b
e=a*b
f=a**2
g=np.sqrt(b)
h=np.log(b)
print("la suma tiene como resultado...", c)
print("la resta tiene como resultado...",d)
print("la multiplicación tiene como resultado...",e)
print("la elevación tiene como resultado...",f)
print("la raiz tiene como resultado...", g)
print("el logaritmo tiene como resultado...",h)
```
