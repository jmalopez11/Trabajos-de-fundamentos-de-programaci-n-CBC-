Un camión circula por una carretera a 20m/s. En 5s. su velocidad pasa a ser de 25m/s. ¿Cual ha sido su aceleración?

Datos

vf=25m/s vi=20m/s t=5s

```python
vf=int(input("ingrese la velocidad final en m/s"))
vi=int(input("ingrese la velocidad inicial en m/s"))
t=int(input("ingrese el tiempo en s:  "))
a=((vf-vi)/t)
print("la resta es: ",vf-vi)
print("El resultado final es de m/s2: ",a)
```
