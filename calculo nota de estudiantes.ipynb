descripción: leer de un estudiante su nombre y las notas de los 3 parciales,
se necesita calcular su nota definitiva sabiendo que es el promedio de los parciales,
además imprimir un concepto de aprobó o reprobó

Autor: grupo FP_dia

""

```python
#Declaración de variables
Var_ent_cad_nombreEstudiante= "NN"
Var_ent_float_notaPP= 0.0
Var_ent_float_notaSP= 0.0
Var_ent_float_notaTP= 0.0
Var_proSal_float_notaDEF= 0.0
Var_proSal_float_cad_concepto= ""
```

```python
#LECTURA DE DATOS-CAPTURA POR TECLADO
Var_ent_cad_nombreEstudiante = input("digite nombre del estudiante");
Var_ent_float_notaPP = float(input("digite nota primer parcial: "))
Var_ent_float_notaSP = float(input("digite nota segundo parcial: "))
Var_ent_float_notaTP = float(input("digite nota tercer parcial: "))
```

```python
# Procesos
#calculo de nota definitiva por promedio
var_proSal_float_notaDefPro=(Var_ent_float_notaPP+Var_ent_float_notaSP+Var_ent_float_notaTP)/3
# cálculo de nota definitiva por porcentaje
Var_proSal_float_notaDEF=(Var_ent_float_notaPP*0.25+Var_ent_float_notaSP*0.25+Var_ent_float_notaTP)
# cálculo del concepto si aprueba
if (var_proSal_float_notaDefPro>=3.0):
  Var_proSal_float_cad_concepto ="Aprobó - te promueves al siguiente curso"
  
# cálculo del concepto si no aprueba
if (var_proSal_float_notaDefPro<=3.0):
  var_proSal_cad_concepto="Reprobó -sigue intentandolo"
```

```python
#salidas o impresiones por pantalla
print("nombre de Estudiante: ",Var_ent_cad_nombreEstudiante)
print("Definitiva por promedio: ",var_proSal_float_notaDefPro)
print("concepto: ",var_proSal_cad_concepto)
print("Definitiva por porcentaje: ",var_proSal_float_notaDefPro)
```
